```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
excel_data = pd.read_excel('Data.xlsx')
excel_data
```

```python
kaz_SSR = excel_data.loc[excel_data["Country"] == "Kazakh SSR"]
kaz_SSR
```

```python
countries = {i for i in excel_data["Country"]}
info_by_countries = dict.fromkeys(countries)
info_by_countries
```

```python
for i in info_by_countries:
    info_by_countries[i] = excel_data.loc[excel_data["Country"] == "Kazakhstan" ]
info_by_countries
```

```python
filt = excel_data[" Life_Expectancy_Person"] >= 72
filt2 = excel_data["Country"] == "Kazakhstan"
save_to_excel = excel_data.where(filt & filt2).dropna()
save_to_excel
```

```python
# --- bar chart or histogram
print(excel_data[" Count_Person"].plot.hist())

plt.xlabel('coefficent')
plt.ylabel('population')
plt.show()


# --- line plots ---
excel_data.plot.line(x="Year", y=" Count_Person_Is_Internet_User_PerCapita")

plt.ylabel('Internet users per capita')
plt.xlabel("Year")
plt.show()


# --- Hex Bin plot ---
excel_data.plot.hexbin(x="Year", y=" Life_Expectancy_Person", gridsize=7, cmap="Reds")
```
